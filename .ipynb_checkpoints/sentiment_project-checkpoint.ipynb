```python
import pandas as pd
import numpy as np
import re

# Scikit-learn
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, RocCurveDisplay

# Matplotlib per i grafici
import matplotlib.pyplot as plt

# Carico il dataset
df = pd.read_csv('Synthetic Reviews.csv')  # o 'synthetic_reviews.csv' a seconda del nome
df.head()
```

```python
# Funzione di clean basic: lowercase + rimozione caratteri non-alfabetici
def clean_text(s):
    s = s.lower()
    s = re.sub(r'[^a-z\s]', '', s)
    return s

df['clean'] = df['text'].apply(clean_text)
df[['text','clean','label']].head()
```

```python
X = df['clean']
y = df['label']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print(f"Train size: {X_train.shape[0]} — Test size: {X_test.shape[0]}")
```

```python
# TF-IDF features (unigram + bigram)
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf  = vectorizer.transform(X_test)

# Naive Bayes
model = MultinomialNB(alpha=1.0)
model.fit(X_train_tfidf, y_train)
```

```python
# Predizioni
y_pred  = model.predict(X_test_tfidf)
y_proba = model.predict_proba(X_test_tfidf)[:,1]

# Classification report
print(classification_report(y_test, y_pred))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix:\n", cm)

# ROC-AUC e curva ROC
auc = roc_auc_score(y_test, y_proba)
print(f"ROC-AUC: {auc:.3f}")
RocCurveDisplay.from_predictions(y_test, y_proba)
plt.title('ROC Curve')
plt.show()
```

```python
feature_names = vectorizer.get_feature_names_out()

# Top 10 token più associati a negativo (classe 0)
neg_ids = np.argsort(model.feature_log_prob_[0])[:10]
# Top 10 token più associati a positivo (classe 1)
pos_ids = np.argsort(model.feature_log_prob_[1])[-10:]

print("Top Negative Tokens:", feature_names[neg_ids])
print("Top Positive Tokens:", feature_names[pos_ids])
```

```python
import pickle

# Salva il vettorizzatore
with open("vectorizer.pkl", "wb") as f:
    pickle.dump(vectorizer, f)

# Salva il modello Naive Bayes
with open("nb_model.pkl", "wb") as f:
    pickle.dump(model, f)

print("Artifacts salvati: vectorizer.pkl, nb_model.pkl")
```
